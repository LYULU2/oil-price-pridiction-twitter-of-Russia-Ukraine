```python
all_review = set() # set ÂéªÈáç
```

```python
def roll_window_to_bottom(browser, stop_length=None, step_length=700):
    """selenium ÊªöÂä®ÂΩìÂâçÈ°µÈù¢ÔºåÂêë‰∏ãÊªë
    :param browser: seleniumÁöÑwebdriver
    :param stop_length: ÊªëÂä®ÁöÑÊúÄÂ§ßÂÄº
    :param step_length: ÊØèÊ¨°ÊªëÂä®ÁöÑÂÄº
    """
    original_top = 0
    while True:  # Âæ™ÁéØÂêë‰∏ãÊªëÂä®
        
        #print(stop_length)
        if stop_length:
            if stop_length - step_length <= 0:
                browser.execute_script("window.scrollBy(0,{})".format(stop_length))
                break
            stop_length -= step_length

        browser.execute_script("window.scrollBy(0,{})".format(step_length))
        time.sleep(0.4+ random.random())  # ÂÅúÈ°ø‰∏Ä‰∏ã
        res = browser.find_elements(By.XPATH,'//article[@class = "css-1dbjc4n r-1loqt21 r-18u37iz r-1ny4l3l r-1udh08x r-1qhn6m8 r-i023vh r-o7ynqc r-6416eg"]')
        for ele in res:
            #print(ele)
            all_review.add(ele.text)
        
        check_height = browser.execute_script(
            "return document.documentElement.scrollTop || window.pageYOffset || document.body.scrollTop;")
        if check_height == original_top:  # Âà§Êñ≠ÊªëÂä®ÂêéË∑ùÈ°∂ÈÉ®ÁöÑË∑ùÁ¶ª‰∏éÊªëÂä®ÂâçË∑ùÈ°∂ÈÉ®ÁöÑË∑ùÁ¶ª
            break
        original_top = check_height
        
```

```python
#stop_length : Âêë‰∏ãÊªöÂä®ÁöÑÂÉèÁ¥†‰ΩçÔºå step‰∏∫ÊØèÊ¨°ÊªöÂ§öÂ∞ëÂÉèÁ¥†
roll_window_to_bottom(browser, stop_length=10000)
```
